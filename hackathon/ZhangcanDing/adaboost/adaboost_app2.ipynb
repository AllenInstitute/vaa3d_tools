```python
import numpy as np
import matplotlib.pyplot as plt
#matplotlib inline
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_gaussian_quantiles

X=[]
y=[]
```

```python
import os 

def crop_app2(path,x0,x1,y0,y1,z0,z1):
    x= path.split(".v3draw")
    manual=x[0]+"_manual.swc"
    commandStr = " /home/dzc/v3d_external/bin/vaa3d " \
                 " -x /home/dzc/v3d_external/bin/plugins/adaboost/libadaboost.so " \
                 " -f func1 " \
                 " -i  {} {} " \
                 " -o {} {} {}" \
                 " -p {} {} {} {} {} {}".format(path, manual, path+"_crop.v3draw", path+"_crop.swc", path+"_shift.swc",x0,x1,y0,y1,z0,z1)
    #"-p {} 0 1".format(path,path+".marker")
    #print(commandStr)
    #os.system(commandStr)
    out=None
    out=float(os.popen(commandStr).readlines()[-2])
    return out
    
```

```python
path="/home/dzc/classification/classification/false_positive/noisy/11192.3_39880.4_3050.26.v3draw"
out=crop_app2(path,32,96,32,96,32,96)
out

scale=[1,32,64,96]
for i in scale:
    for j in scale:
        for k in scale:
            X.append([k,k+31,j,j+31,i,i+31])
            dist=crop_app2(path,k,k+31,j,j+31,i,i+31)
            if dist>0 and dist<3:
                y.append(1)
            else:
                y.append(-1)

scale1=[1,32,64]
for i in scale1:
    for j in scale1:
        for k in scale1:
            X.append([k,k+63,j,j+63,i,i+63])
            dist=crop_app2(path,k,k+63,j,j+63,i,i+63)
            if dist>0 and dist<3:
                y.append(1)
            else:
                y.append(-1)

                
            
```

```python
bdt = AdaBoostClassifier(DecisionTreeClassifier(max_depth=2, min_samples_split=20, min_samples_leaf=5),
                         algorithm="SAMME",
                         n_estimators=200, learning_rate=0.8)

bdt.fit(X, y)
```
