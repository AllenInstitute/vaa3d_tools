```python
import SimpleITK as sitk
import random


def load_itk(filename):
    itkimage= sitk.ReadImage(filename)
    # Convert the image to a  numpy array first and then shuffle the dimensions to get axis in the order z,y,x
    scan= sitk.GetArrayFromImage(itkimage)
    #origin= np.array(list(reversed(itkimage.GetOrigin())))
    # Read the spacing along each dimension
    #spacing= np.array(list(reversed(itkimage.GetSpacing())))
    return scan 

scan0=load_itk("/home/dzc/anaconda/testimg/11958.2_39841.4_3433.28.tif")
scan1=load_itk("/home/dzc/anaconda/testimg/adatest.tif")
X=[]
y=[]
for i in range(scan0.shape[0]):
    for j in range(scan0.shape[1]):
        for k in range(scan0.shape[2]):
            if scan1[i,j,k]==255:
                X.append([scan0[i,j,k]])
                y.append(255)
            
print(len(y))
for a in range(len(X)):
    i=random.randint(1,scan1.shape[0]-2)
    j=random.randint(1,scan1.shape[1]-2)
    k=random.randint(1,scan1.shape[2]-2)
    if scan1[i,j,k]==0:
        X.append([scan0[i,j,k]])
        y.append(0)
print(len(y))


#scan0, scan1
#X=scan0.flatten()
#y=scan1.flatten()
#cnt=0
#for i in range(y.shape[0]):
    #if y[i]==255:
        #cnt=cnt+1
```

```python
bdt = AdaBoostClassifier(DecisionTreeClassifier(max_depth=2, min_samples_split=20, min_samples_leaf=5),
                         algorithm="SAMME",
                         n_estimators=200, learning_rate=0.8)

bdt.fit(X, y)
```

```python
scan2=load_itk("/home/dzc/anaconda/testimg/11958.2_39841.4_3433.28.tif")
#scan2.flatten()
#scan2.reshape(-1,1)

pre=[]
for i in range(scan0.shape[0]):
    for j in range(scan0.shape[1]):
        for k in range(scan0.shape[2]):
            pre.append([scan2[i,j,k]])
pre
```

```python
mask_array=np.array(bdt.predict(pre))
#print(scan2.size())

#np.reshape(mask_array,128,(128,128))
mask_img=sitk.GetImageFromArray(mask_array.reshape(128,128,128))
mask_img_conv=sitk.Cast(mask_img,sitk.sitkUInt8)
sitk.WriteImage(mask_img_conv,"/home/dzc/anaconda/testimg/adatest_mask.tif" )
```
