```python
fnames, abs_paths = get_fnames_and_abspath_from_dir(CUBE_NPY_DIR)

# fetch first tiff
test_npy = abs_paths[0]
print("file: ", test_npy)

import numpy as np
test_npy = np.load(test_npy)
import matplotlib

matplotlib.use('TkAgg')
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
# Set up figure & 3D axis for animation
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.patch.set_alpha(0.5) # Set semi-opacity


# # prepare some coordinates
# x, y, z = np.indices((8, 8, 8))

# # draw cuboids in the top left and bottom right corners, and a link between them
# cube1 = (x < 3) & (y < 3) & (z < 3)
# cube2 = (x >= 5) & (y >= 5) & (z >= 5)
# link = abs(x - y) + abs(y - z) + abs(z - x) <= 2
# #print(link)

# # combine the objects into a single boolean array
# voxels = cube1 | cube2 | link

# set the colors of each object
# colors = np.empty(voxels.shape, dtype=object)
# colors[link] = 'red'  # link overlaps with cubes, but we print cubes after to superimpose
# colors[cube1] = 'blue'
# colors[cube2] = 'green'

def normalize(arr):
    arr_min = np.min(arr)
    return (arr - arr_min) / (np.max(arr) - arr_min)


def expand_coordinates(indices):
    x, y, z = indices
    x[1::2, :, :] += 1
    y[:, 1::2, :] += 1
    z[:, :, 1::2] += 1
    return x, y, z

def explode(data):
    shape_arr = np.array(data.shape)
    size = shape_arr[:3] * 2 - 1
    exploded = np.zeros(np.concatenate([size, shape_arr[3:]]), dtype=data.dtype)
    exploded[::2, ::2, ::2] = data
    return exploded


def plot_cube(cube, angle=320):
    cube = normalize(cube)
    print(cube)

    facecolors = cm.viridis(cube)
    facecolors[:, :, :, -1] = cube
    facecolors = explode(facecolors)

    filled = facecolors[:, :, :, -1] != 0
    x, y, z = expand_coordinates(np.indices(np.array(filled.shape) + 1))

    fig = plt.figure(figsize=(30 / 2.54, 30 / 2.54))
    ax = fig.gca(projection='3d')
    ax.axis('off')
    ax.view_init(30, angle)

    ax.voxels(x, y, z, filled, facecolors=facecolors)
    plt.show()

IMG_DIM = 16
plot_cube(test_npy)

# and plot everything
# ax.voxels(test_npy, edgecolor='k', alpha=0,
#         linewidth=0.5) # , facecolors=colors, 

#ax.set_alpha(0.5)

# plt.show()
```
